```python
import pandas as pd
import numpy as np

import wbgapi as wb

import warnings
warnings.filterwarnings('ignore')
```

```python
# Importing the World Bank data using API
   
country = 'JPN' #Japan

#Defining the time range. Ignoring years 2024 and 2025 as consumption data is not available.
years = range(1974,2023)

# GDP Data
gdp_data = wb.data.DataFrame('NY.GDP.PCAP.CD',country,time = years)

# Total Consumption Data
total_consumption = wb.data.DataFrame('NE.CON.PRVT.CD',country, time = years)

# Population Data
population = wb.data.DataFrame('SP.POP.TOTL',country, time = years)
```

```python
# Calculating per capita consumption
consumption_per_capita = total_consumption/population
```

```python
gdp_data.head()
```

```python
# Creating a dataframe
df = pd.DataFrame (
    {'GDP_per_capita':gdp_data.iloc[0],
    'Consumption_per_capita':consumption_per_capita.iloc[0]
    }
)
```

```python
df.head(60)
```

```python
df = df.reset_index()
df.head()
```

```python
df.columns = ['Year','GDP_per_capita','Consumption_per_capita']
df['Year'] = df['Year'].astype(str).str.replace("YR","")
df = df.set_index('Year')
```
